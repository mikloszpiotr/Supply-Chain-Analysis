```python
#import libraries
import pandas as pd
import pygal
```

```python
#import data set, dataset was downloaded from Kaggle:https://www.kaggle.com/datasets/hamidrezabakhtaki/fmcg-export-sale
df=pd.read_csv("ExportSales.csv",index_col=1,parse_dates=True)
df.head(2)
```

```python
#Sort index to be sure that dates are increasing, change data sequence to monthly basis,check if index is correct
df=df.sort_index(ascending=True)
df=df.resample('M').sum()
df.index
```

```python
#Delete missing values
df = df.dropna(axis=0, how='any')
```

```python
# Create a line chart for quantity
line_chart = pygal.Line(x_label_rotation=90)
line_chart.title = 'QTY by Month'
line_chart.x_labels = df.index
line_chart.add('QTY', df["QTY"])
```

```python
# Create a line chart for Total Price
line_chart = pygal.Line(x_label_rotation=90)
line_chart.title = 'Total Price by Month'
line_chart.x_labels = df.index
line_chart.add('Total Price', df["Total Price"])
```

```python
#Create a bar chart for Quantity
bar_chart = pygal.Bar()
bar_chart.title = 'Category QTY'
bar_chart.add("QTY", df["QTY"])
```

```python
#Create a bar chart for Total Price
bar_chart = pygal.Bar()
bar_chart.title = 'Category Total Price'
bar_chart.add("Total Price", df["Total Price"])
```
