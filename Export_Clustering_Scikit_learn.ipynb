```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
#import data set, dataset was downloaded from Kaggle:https://www.kaggle.com/datasets/hamidrezabakhtaki/fmcg-export-sale
df=pd.read_csv("ExportSales.csv")
df.head(2)
```

```python
#Define a size
plt.rcParams["figure.figsize"] = (12, 8)
```

```python
#Filter data for clustering algorithm
df_cl = df[["Category","QTY"]]
```

```python
# Reshape the 1D array to a 2D array
X = df_cl["QTY"].to_numpy().reshape(-1, 1)
```

```python
# Perform KNN clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
# Plot the data points and cluster centers
plt.scatter(df_cl['Category'],df_cl['QTY'], c=kmeans.labels_, cmap='rainbow')
plt.title('KNN Clustering with K=3')
plt.xlabel('Category')
plt.ylabel('QTY')
plt.show()
```

```python
# Perform Agglomerative clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_clustering.fit(X)
```

```python
# Plot the data points and cluster centers
plt.scatter(df_cl['Category'],df_cl['QTY'], c=agg_clustering.labels_, cmap='rainbow')
plt.title('KNN Clustering with K=3')
plt.xlabel('Category')
plt.ylabel('QTY')
plt.show()
```
